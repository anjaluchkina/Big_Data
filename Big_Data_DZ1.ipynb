```python
import json
from functools import reduce
from pathlib import Path

def mapper(path):
    with open(path, 'r') as f:
        info = json.load(f)
    score = float(info['movieIMDbRating'])
    return (1, score, score*score)  # Возвращаем кортеж с количеством оценок, суммой оценок и суммой квадратов оценок

def reducer(score_data1, score_data2):
    if score_data1 is None:  # проверка пустого входного списка
        return score_data2
    elif score_data2 is None:
        return score_data1
    else:
        n1, sum1, sum_sq1 = score_data1
        n2, sum2, sum_sq2 = score_data2

        n = n1 + n2
        mean = (sum1 + sum2) / n
        M2 = sum_sq1 + sum_sq2 + (n1 * n2 / n) * ((sum1 / n1) - (sum2 / n2)) ** 2

        return n, mean, M2

scores = map(mapper, Path('imdb-user-reviews').glob('**/*'))
result = reduce(reducer, scores, None)

if result is not None:
    n, mean, M2 = result
    print(mean, (M2 / n) ** (1/2))
else:
    print("No data to process.")

```
